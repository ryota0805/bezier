```python
import numpy as np
from scipy.interpolate import splev, splrep
import matplotlib.pyplot as plt
```

```python
# サンプルのx座標とy座標
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 3, 1, 2, 2, 0])

# Spline曲線を計算
tck = splrep(x, y, k=3, s=0)

# Spline曲線を評価
x_eval = np.linspace(0, 5, 1000)
y_eval = splev(x_eval, tck)
```

```python
plt.plot(x_eval, y_eval)
```

```python
# B-Spline曲線の1階微分を計算
dx_dt = splev(x_eval, tck, der=1)
dy_dt = splev(x_eval, tck, der=1)
```

```python
plt.plot(y_eval)
plt.show()
plt.plot(dy_dt)
plt.show()
```

```python
import numpy as np
from scipy.interpolate import splprep, splev
import matplotlib.pyplot as plt

# サンプルのx座標とy座標
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 3, 1, 2, 2, 0])

# splprepを使用してB-Spline曲線を生成し、制御点を取得
tck, u = splprep([x, y], k=3, s=0)

# B-Spline曲線を評価
x_eval, y_eval = splev(np.linspace(0, 1, 1000), tck)

# 1階微分を生成
dx_dt, dy_dt = splev(np.linspace(0, 1, 1000), tck, der=1)

# 2階微分を生成
d2x_dt2, d2y_dt2 = splev(np.linspace(0, 1, 1000), tck, der=2)

plt.plot(y_eval)
```
